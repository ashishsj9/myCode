```python
import csv

with open ('/Users/ashishjhade/documents/techLearning/weight-height.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter = ',')
    ctr = 0
    for row in csv_reader:
        if ctr == 0:
            print("Please find the details below")
            ctr = ctr+1
        else:    
            gender = row[0]
            height = round(float(row[1]), 0)
            weight = round(float(row[2]),0)
            print ("Gender is: "+str(gender)+", Height is: "+str(height)+" inchs, Weight is: "+str(weight)+" Lbs")
```
